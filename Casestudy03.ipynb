# 1.First of all, check whether there is any relationship between the amount spent for advertising through five different companies and profit generated

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Add_sales_data_2018.csv')
```

```python
data.head()
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
corrmatrix=df.corr()
plt.subplots(figsize=(7,3))
sns.heatmap(corrmatrix,vmin=-1,vmax=1,annot=True,linewidths=0.2)
```

There is a high degree of positive correlation between the amount spent for advertising through five different companies and  the profit generated from it.Among the 5 companies,company 2 is highly correlated among the lot with an ideal correlation value of 1 and the remaining companies have a correlation of 0.99 which is super close to the ideal value.

# 2.Build a regression model to predict the profit from the given data and comment your findings on it

```python
y=data['Profit(in Thousands)']
X=data.drop(['Profit(in Thousands)','Date'],axis=1)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.30)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
```

```python
model=lr.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
print(y_test)
```

```python
print(predictions)
```

Comparing the input and output,it is evident that the predictions made by the linear regression model are very accurate.This means that the values are highly correlated and is close to the ideal correlation.
